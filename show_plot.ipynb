```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('follow_relationship.csv', nrows=100)
df = pd.DataFrame({'from': data['user_id'], 'to': data['follower_id']})
print(data)
```

```python
G = nx.from_pandas_edgelist(df, 'from', 'to', create_using=nx.DiGraph())

# Make the graph
nx.draw(G, with_labels=True, node_size=300, alpha=0.5, arrows=True)
plt.title("Directed Plot")
plt.show()
```
